# Property Model

This document was constructed analyze specific properties in Nashville, Tennessee. The data was scraped from Redfin via another script. The ultimate objective is to identify various features and build a machine Learning model to determine prices based on features.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
file=r'C:\Users\claudius.ibine\Downloads\redfin_2018-05-31-06-19-11.csv'
df=pd.read_csv(file)
df=df[['PRICE','BEDS','BATHS','YEAR BUILT','SQUARE FEET','DAYS ON MARKET']]
```

```python
df.head(16)
```

```python
df['YEAR BUILT']=df['YEAR BUILT'].fillna(df['YEAR BUILT'].mode()[0])
df['DAYS ON MARKET']=df['DAYS ON MARKET'].fillna(df['DAYS ON MARKET'].mode()[0])
df['YEAR BUILT']=df['YEAR BUILT'].apply(np.int64)
df['DAYS ON MARKET']=df['DAYS ON MARKET'].apply(np.int64)
```

```python
x=df[['BEDS','BATHS','YEAR BUILT','SQUARE FEET','DAYS ON MARKET']]
y=df['PRICE'].values.tolist()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#build dictionary to store model scores
scores={'KNN':0,'DTC':0,'LR':0,'GNB':0,'RF':0,'SVM':0,'LDA':0}

knn=KNeighborsClassifier(n_neighbors=3)
knn=knn.fit(x_train,y_train)
knn_accuracy=knn.score(x_test,y_test)
scores['KNN']=knn_accuracy
print('KNN accuracy score is: ' + str(knn_accuracy))

dtc=tree.DecisionTreeClassifier()
dtc=dtc.fit(x_train,y_train)
dtc_accuracy=dtc.score(x_test,y_test)
scores['DTC']=dtc_accuracy
print('DTC accuracy score is: ' + str(dtc_accuracy))

lr=LogisticRegression()
lr=lr.fit(x_train,y_train)
lr_accuracy=lr.score(x_test,y_test)
scores['LR']=lr_accuracy
print('LR accuracy score is: ' + str(lr_accuracy))

gnb=GaussianNB()
gnb=gnb.fit(x_train,y_train)
gnb_accuracy=gnb.score(x_test,y_test)
scores['GNB']=gnb_accuracy
print('GNB accuracy score is: ' + str(gnb_accuracy))

rf=RandomForestClassifier(n_estimators=1000)
rf=rf.fit(x_train,y_train)
rf_accuracy=rf.score(x_test,y_test)
scores['RF']=rf_accuracy
print('RF accuracy score is: ' + str(rf_accuracy))

svm=SVC()
svm=svm.fit(x_train,y_train)
svm_accuracy=svm.score(x_test,y_test)
scores['SVM']=svm_accuracy
print('SVM accuracy score is: ' + str(svm_accuracy))

lda=LinearDiscriminantAnalysis()
lda=lda.fit(x_train,y_train)
lda_accuracy=lda.score(x_test,y_test)
scores['LDA']=lda_accuracy
print('LDA accuracy score is: ' + str(lda_accuracy))
```

```python
lda=LinearDiscriminantAnalysis()
lda=lda.fit(x_train,y_train)
accuracy=lda.score(x_test,y_test)
lda_prediction=lda.predict(x_test)
```

```python
df=x_test.reset_index()
df=df[['BEDS','BATHS','YEAR BUILT','SQUARE FEET','DAYS ON MARKET']]
actual=(list(y_test))
prediction=(list(lda_prediction))
df['actual']=pd.Series(actual)
df['prediction']=pd.Series(prediction)
df
```
